## 1. Intoduction: 

### 1.1 Business Problem


More than 38,000 people die every year in crashes on U.S. roadways. The U.S. traffic fatality rate is 12.4 deaths per 100,000 inhabitants. An additional 4.4 million are injured seriously enough to require medical attention. Road crashes are the leading cause of death in the U.S. for people aged 1-54.

Road accidents are increasing day to day all over the world due to several reasons. These accidents may happen due to either avoidable circumstances or un-avoidable circumstances.

- Avoidable scenarios: Speeding, not wearing safety gear such as helmets or seat belts, breaking traffic rules.
- Un avoidable scenarios: Bad roads, weather conditions, lighting conditions, traffic conditions.

While avoidable accident cases could be drastically reduced by enforcing harsh/strict penalties by law. Although, un-avoidable accident situations are difficult to control, chances of fatalities or severe injuries can be reduced by notifying local police authorities, healthcare systems, traffic authorities, safety personnel in advance when there is a high probability of accidents so that they can make better decisions in avoiding the accidents or respond quickly in case of any accidents.

At the end of this project, we will develop a machine learning model based on several differrnt features which will predict the severity of the accident. This model will aid stake holders to be proactive and take impactful decisions to reduce chances/severity of the accident.

### Weather condition attribute:

According to Road Weather Management Program, most of weather-related crashes happen on wet-pavement and during rainfall. Also, Winter-condition and fog are another two main reasons for weather-related accidents.

```python
print("Unique number of values: ",df_data["Weather_Condition"].nunique())
print("Unique values: ",df_data["Weather_Condition"].unique().tolist())
```

Since, there are plenty of unique values in Weather Conndition (n=127). Let's simplify and create features for most common weather condtions and drop 'Weather_Condtion' attribute:

```python
df_data['Clear']=np.where(df_data['Weather_Condition'].str.contains('Clear', case=False, na=False), 1,0)
df_data['Rain']=np.where(df_data['Weather_Condition'].str.contains('Rain|Drizzle|storm|Light Rain Showers', case=False, na=False), 1,0)
df_data['Heavy_Rain']=np.where(df_data['Weather_Condition'].str.contains('Heavy Rain|Heavy Drizzle|Thunderstorm|Rain Showers|Heavy Thunderstorms and Rain', 
                                                                         case=False, na=False), 1,0)
df_data['Cloud']=np.where(df_data['Weather_Condition'].str.contains('Cloud|Overcast', case=False, na=False), 1,0)
df_data['Heavy_Snow'] = np.where(df_data['Weather_Condition'].str.contains('Heavy Snow|Heavy Sleet|Heavy Ice Pellets|Snow Showers|Squalls', case=False, na = False), 1, 0)
df_data['Fog'] = np.where(df_data['Weather_Condition'].str.contains('Fog', case=False, na = False), 1, 0)

# We will assign NA to newly created weather conditions where 'Weather_Condtion' is null
weather=['Clear','Rain','Heavy_Rain','Cloud','Heavy_Snow','Fog']
for i in weather:
    df_data.loc[df_data['Weather_Condition'].isnull(),i] = df_data.loc[df_data['Weather_Condition'].isnull(),'Weather_Condition']

#df_data.loc[:,['Weather_Condition'] + weather]

df_data.drop(['Weather_Condition'], axis=1, inplace=True)
    
```

### Resampling the data:

We will now look at distribution of severity cases:

```python
df_data['Severity'].value_counts()
```

We will focus more on Severity level 4 since it represents more severe/fatal accidental cases.Therefore, we will compare the total number of Severity level 4 accident cases against the rest.

```python
df_data['Severity_4'] =0
df_data.loc[df_data['Severity']==4, 'Severity_4'] =1
df_data.Severity_4.value_counts()
```

From above, we notice that the data is unbalanced. So it would not accurate results with using unbalanced data. Therefore, in order to address this issue, we will perform over and undersampling. We will randomly oversampled level 4 to 100000 and other levels will be randomly undersampled to 100000. 

```python
df_data = df_data.drop(['Severity'], axis = 1)
df_bal = pd.concat([df_data[df_data['Severity_4']==1].sample(100000, replace = True),
                   df_data[df_data['Severity_4']==0].sample(100000)], axis=0)
print('Resampled data:') 
df_bal.Severity_4.value_counts()
```

Let us normalize values with extremely skewed distribution:

```python
# importing module
from scipy.stats import boxcox


df_data['Pressure_bc']= boxcox(df_data['Pressure(in)'].apply(lambda x: x+1),lmbda=6)
df_data['Visibility_bc']= boxcox(df_data['Visibility(mi)'].apply(lambda x: x+1),lmbda = 0.1)
df_data['Wind_Speed_bc']= boxcox(df_data['Wind_Speed(mph)'].apply(lambda x: x+1),lmbda=-0.2)
df_data = df_data.drop(['Pressure(in)','Visibility(mi)','Wind_Speed(mph)'], axis=1)
```

```python
df_data = df_data.replace([True, False], [1,0])

category = ['Side','State','Timezone','Wind_Direction', 'Weekday', 'Month', 'Hour']
df_data[category] = df_data[category].astype('category')
df_data = pd.get_dummies(df_data, columns=category, drop_first=True)

df_data.head()
```

### 4.2 Model Building

#### Train/Test Split

We shall resample and split data into X and y.

We will standardize the value and split data into X_train, X_test, y_train, y_test.

```python
df_data.drop('Start_Time', axis=1, inplace=True)

df_bal = pd.concat([df_data[df_data['Severity_4']==1].sample(100000, replace = True), 
                   df_data[df_data['Severity_4']==0].sample(100000)], axis=0)
# split X, y
X = df_bal.drop('Severity_4', axis=1)
y= df_bal['Severity_4']

# Standardizing the features based on unit variance
from sklearn.preprocessing import StandardScaler

X = StandardScaler().fit_transform(X)

# split train test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=3)
```

### 4.2.1 Logistic Regression:

```python
# Logistic regression with default setting.
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(C=0.01, solver='liblinear', max_iter=10000).fit(X_train, y_train)

accuracy_train = clf.score(X_train, y_train)
print("Train Accuracy: %.1f%%"% (accuracy_train*100))
accuracy_test = clf.score(X_test,y_test)
print("Test Accuracy: %.1f%%"% (accuracy_test*100))
```

### Jaccardian Similarity index:

```python
yhat = clf.predict(X_test)

#Jaccardian Similarity index

from sklearn.metrics import jaccard_score
jaccard_score(y_test, yhat)
```

### Confusion matrix

```python
from sklearn.metrics import multilabel_confusion_matrix

confusion_mat = multilabel_confusion_matrix(y_true=y_test, y_pred=yhat,
                           labels=[1])

conf_matrix = pd.DataFrame(data=confusion_mat[0],
                           columns=['Predicted:0','Predicted:1'],index=['Actual:0','Actual:1'])
plt.figure(figsize = (8,5))
sns.heatmap(conf_matrix, annot=True,fmt='d',cmap="YlGnBu").set_title(
    "Confusion Matrix (resampled data)\n-Default Logistic Regression", fontsize=16)
plt.show()
```

### 4.2.2 Decision Tree Classifier:

In Decision Tree Classifier, we achieve maximum accuracy with maximum sample split size of 3 with train accuracy of 99.4% and Test accuracy of 81.0%.

```python
from sklearn import tree
# Training step, on X_train with y_train
tree_clf = tree.DecisionTreeClassifier(min_samples_split = 3)
tree_clf = tree_clf.fit(X_train,y_train)

tree_accuracy_train = tree_clf.score(X_train, y_train)
print("Train Accuracy: %.1f%%"% (tree_accuracy_train*100))
tree_accuracy_test = tree_clf.score(X_test,y_test)
print("Test Accuracy: %.1f%%"% (tree_accuracy_test*100))
```

```python
from sklearn.metrics import confusion_matrix
tree_yhat = tree_clf.predict(X_test)

confusemat = confusion_matrix(y_true=y_test, y_pred=tree_yhat)

conf_matrix = pd.DataFrame(data=confusemat,
                           columns=['Predicted:0','Predicted:1'],index=['Actual:0','Actual:1'])
plt.figure(figsize = (8,5))
sns.heatmap(conf_matrix, annot=True,fmt='d',cmap="YlGnBu").set_title("Confusion Matrix (resampled data)\n Decision Tree", fontsize=16)
plt.show()
```

```python
importance = pd.DataFrame(np.zeros((X_train.shape[1], 1)), columns=['importance'], 
                           index=df_data.drop('Severity_4',axis=1).columns)

importance.iloc[:,0] = tree_clf.feature_importances_

importance.sort_values(by='importance', inplace=True, ascending=False)
importance25 = importance.head(25)

plt.figure(figsize=(10, 10))
sns.barplot(x='importance', y=importance25.index, data=importance25)

plt.xlabel('')
plt.tick_params(axis='x', labelsize=10)
plt.tick_params(axis='y', labelsize=10)
plt.title('Decision Tree Classifier Feature Importance', size=15)

plt.show()
```
